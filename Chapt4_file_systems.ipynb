### 4.1.2 File Structure ###
There are 3 common models.

#### Byte Sequence####
1. The operating system regards files as nothing more than byte sequences. It provides the maximum amount of flexibility. Users can put anything they want in thier files.

2. All versions of UNIX and Windows use this model.

#### Record Sequenc####
1. A file is a sequence of fixed-length records, each with some internal structure.

2. The central idea is that the read operation returns one record and the write operation overwrites or appends one record.

3. No current general-purpose system uses this model as its primary file system any more. But it was a common model on mainframe computers.

#### Tree ####
1. A file consists of a tree of records, not necessarily all the same length, each containing a **key** field in a fixed position in the record. The tree is sorted on the key field, to allow rapid searching for a particular key.

2. The basic operation is to get the record with a specific key.

3. Used on some large mainframe computers for commercial data processing.

### 4.1.3 File Types ###
1. Regular files: contain user information.

2. Directories

3. Character speical files: related to input/output and used to model serial I/O devices, such as terminals, printers, and networks.

4. Block special files: sued to model disks.

5. Although technicially the file is just a sequence of bytes, the operating system will execute a file only if it has the proper format. It has five sections: header, text, data, relocation bits, and symbol table.

### 4.1.4 File Access ###
1. Sequential access

2. Random-access 

3. Two methods for specifying where to start reading:
    - `read` operation gives the position in teh file to start reading at
    - `seek` operation sets the current position. After than, the file can be read sequentially from the now-current position.

### 4.1.5 File Attributes ###
1. A file's attributes, or metadata: other information associated with each file in operating systems, e.g. date and tiem the file was last modified, file size, protection, creator, key position, key length.

## 4.3 File-System Implementations ##

### 4.3.1 File-System Layout ###
1. File sytems are stored on disks. Most disks can be divided up into one or more partitions, with independent file systems on each partition. Sector 0 of the disk is called the ***MBR (Master Boot Record)***.

2. Disk partition: Boot block; superblock; free space mgmt; i-nodes; root dir; files and directories

### 4.3.2 Implementing Files ###
#### Contiguous Allocation ####
1. Store each file as a contiguous run of disk blocks.

2. Advantages:
    - Simple to implement.
    - The read performance is execellent because the entire file can be read from the disk in a single operation.

3. Disadvantages:
    - Over the course of time, the disk becomes fragmented.

#### Linked-List Allocation ####
1. Advantages:
    - Sequantial access is straightfoward.

2. Disadvantages:
    - Random access is slow
    - Generates overhead (i.e. info like pointer words)

#### Linked-List Allocation Using a Table in Memory ####
1. Advantages:
    - Reduces pointer words overhead

2. Disadvantages:
    - The whole table must be in memory all the time to make it work.

#### I-nodese ####
1. A.K.A index-node, which lists the attributes and disk addresses of thefile's blocks.

2. Advantages:
    - The i-node need be in memory only when the corresponding file is open.
